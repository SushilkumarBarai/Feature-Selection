```python
#importing the libraries
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
```

```python

df = pd.read_csv('wine_original.csv')

# first 5 rows of the data
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
X=df.iloc[:,0:13]
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)

from sklearn.ensemble import RandomForestClassifier
```

```python
# only 6 feature
forward_feature_selection=sfs(RandomForestClassifier(n_jobs=-1),
                              k_features=6,
                              forward=True,
                              floating=False,
                              verbose=2,
                              scoring='accuracy',
                              cv=5
                              ).fit(X_train,y_train)



```

```python
forward_feature_selection.k_feature_idx_
```

```python
forward_feature_selection.k_feature_names_
```

```python
forward_feature_selection.k_score_
```

```python
pd.DataFrame.from_dict(forward_feature_selection.get_metric_dict()).T
```
